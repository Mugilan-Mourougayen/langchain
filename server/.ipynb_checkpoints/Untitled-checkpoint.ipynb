```python
from dotenv import load_dotenv
from flask import Flask, request, send_from_directory,send_file,jsonify
from flask_cors import CORS
load_dotenv()
import sys
from langchain.agents import create_csv_agent
from langchain.llms import OpenAI 
```

```python
from langchain.memory import ConversationBufferMemory
memory = ConversationBufferMemory(return_messages=True)
```

```python
agent = create_csv_agent(OpenAI(temperature=0), 
                         'personal.csv', 
                         verbose=True,
                        memory=ConversationBufferMemory())

```

```python

question = """
What is the UL Flexible Metal Conduith?
"""

prompt = f"""
Search the dataset for the UL standard name related to text.
Return the entire row as it is, you dont adjust or modify the row Make sure the entire row is returned as answer excluding the row number start from Standard Title with proper spacing and remove unwanted chracters.
Text: `{question}`
"""
```
